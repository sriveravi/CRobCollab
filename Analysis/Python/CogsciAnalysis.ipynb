```python
def to_number(s):
    try:
        s1 = float(s)
        return s1
    except ValueError:
        return s
```

```python
''' want to group data to do some stats in R

    experiment 1 focuses on:
        mixDataByTrial.txt
        visDataByTrial.txt
        audDataByTrial.txt
        
        
        Unimodal auditory:
            same trials (A1 -> A1) 
            new auditory (A1 -> A2)
        Unimodal visual 
            same trials (V1 -> V1)
            new visual (V1 -> V2)
        Cross-modal was broken into 
            same (A1V1 -> A1V1), 
            new auditory (A1V1 -> A2V1)
            new visual  (A1V1 -> A1V2)
            and different (A1V1 -> A2V2)
 
    ignored all same and different trials and 
    only looked at RTs for new auditory in unimodal and cross-modal conditions and 
    new visual in unimodal and cross-modal conditions. 
    
    Keep only correct trials, and subtract 2 s from each trial (already done for logRT).

'''
files =  ['visDataByTrial.txt', 'mixDataByTrial.txt','audDataByTrial.txt'] 
conds = [ 'vis', 'mix', 'aud'] 

df = pd.DataFrame()
for idx,f in enumerate(files):
    dfTemp = pd.read_csv( f,sep='\t')
    dfTemp['condition'] = conds[idx]
    df = pd.concat( [df,dfTemp ], ignore_index=True)

print df.columns
```
